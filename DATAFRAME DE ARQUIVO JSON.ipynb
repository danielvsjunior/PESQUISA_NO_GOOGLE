```python
import json
```

```python
import pandas as pd
```

```python
import csv
```

```python
with open('arquivos/Cursos.json','r') as arquivo:
    texto = arquivo.read()
    data = json.loads(texto)
```

```python
linhas=[]
for indx in range(len(data)):
    url=data[indx]['url']
    uuid = data[indx]['uuid']
    codigo = data[indx]['codigo']
    descricao = data[indx]['descricao']
    diretoria = data[indx]['diretoria']
    natureza_participacao = data[indx]['natureza_participacao']
    eixo = data[indx]['eixo']
    modalidade = data[indx]['modalidade']
    resolucao_criacao = data[indx]['resolucao_criacao']
    coordenador = data[indx]['coordenador']
    ch = data[indx]['ch']
    matriz_vingente_url = data[indx]['matriz_vingente']['url']
    matriz_vingente_uuid = data[indx]['matriz_vingente']['uuid']
    for indx2 in range(len(data[indx]['matrizes'])):
        url_matrizes = data[indx]['matrizes'][indx2]
        linha_interna=[url,uuid,codigo,descricao,diretoria,natureza_participacao,eixo,modalidade,resolucao_criacao,coordenador,ch,matriz_vingente_url, matriz_vingente_uuid, url_matrizes]
        linhas.append(linha_interna)
```

```python
df=pd.DataFrame(linhas, columns = ('url','id','Código','Descrição','Diretoria','Natureza Participação','Eixo','Modalidade','Resolução Criação','Coordenador','Ch','Matriz Vingente url', 'Matriz Vingente uuid', 'url Matrizes'))
```
